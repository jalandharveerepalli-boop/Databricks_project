### **Initializing Parameters**

```python
dbutils.widgets.text("File_Name" ,"orders")
```

```python
p_var_value = dbutils.widgets.get("File_Name")
p_var_value
```

### **Import Libraries**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

### **Data Reading**

```python
df = spark.readStream.format("cloudFiles")\
          .option("cloudFiles.format", "parquet")\
          .option("cloudFiles.schemaLocation",f"abfss://bronze@datalakete.dfs.core.windows.net/checkpoint_{p_var_value}")\
          .load(f"abfss://source@datalakete.dfs.core.windows.net/{p_var_value}")
```

### **Data Writing**

```python
df.writeStream.format("parquet")\
  .outputMode("append")\
  .option("checkpointLocation", f"abfss://bronze@datalakete.dfs.core.windows.net/checkpoint_{p_var_value}")\
  .option("path", f"abfss://bronze@datalakete.dfs.core.windows.net/{p_var_value}")\
  .trigger(once=True)\
  .start()
```
